## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
import pandas as pd
import numpy as np

# Load Titanic dataset
titanic_df = pd.read_csv("train_titanic.csv")
# Step 1: Replace missing values in 'Embarked' with 'S'
titanic_df['Embarked'].fillna('S', inplace=True)




```

```python
#Impute the missing values in Fare column with median value
titanic_df['Fare'].fillna(titanic_df['Fare'].median(), inplace=True)

```

```python
#Impute the Age column. It has too many missing values but it may be important. 
# Step 3: Impute missing values in 'Age' using normal distribution
age_mean = titanic_df['Age'].mean()
age_std = titanic_df['Age'].std()



#We fill the missing values with numbers randomly drawn from distribution
# Generate random ages for missing values
missing_age_count = titanic_df['Age'].isnull().sum()
random_ages = np.random.normal(age_mean, age_std, missing_age_count)

#that has same mean and standard deviation as non missing Age data
# Fill missing Age values with these random numbers
titanic_df.loc[titanic_df['Age'].isnull(), 'Age'] = random_ages

# Save the cleaned dataset
titanic_df.to_csv("train_titanic_cleaned.csv", index=False)

# Show summary of missing values after imputation
print("Missing values after imputation:\n", titanic_df.isnull().sum())


```
